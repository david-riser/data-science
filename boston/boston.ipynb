# Boston Housing Data Analysis 
This notebook presents a basic analysis of the Boston housing data.  It makes use of several common libraries in data science.  The following topics are presented: 
- Loading and naming data in a pandas DataFrame
- Getting to know the data by
    - looking at basic statistics
    - visualization
- Using basic classifiers to predict the price of houses based on a few of the important features selected above.

```python
# basics used throughout the notebook 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline 
plt.rc('font', family='serif')
plt.rc('font', size=18)
```

## Loading and naming data in a pandas DataFrame

```python
from sklearn.datasets import load_boston
dataset = load_boston()
dataset['data'].shape
```

```python
print(dataset['DESCR'])
```

```python
data = pd.DataFrame(data=dataset['data'], columns=dataset['feature_names'])
data['price'] = pd.Series(dataset['target'], index=data.index)
```

```python
data.drop(['B'], axis=1, inplace=True)
```

```python
#naming_convention = {'CRIM':'crime', 'RM':'rooms','AGE':'age','DIS':'distance','TAX':'tax', 'PTRATIO':'median_value'}
#data.rename(index=str, columns=naming_convention, inplace=True)
```

### Getting to Know the Data 
First we check the limits, mean, and deviation of each variable.  Then we look at the correlation between each variable and the sell price of the house (identifying those which may be good predictors).  

Then, plotting of the distributions of each variable are created.  Finally, a scatterplot is created for each variable with the sale price.  This visualization allows us to see the correlations that were found above.

```python
data.describe()
```

```python
data.corr()['price']
```

From this list it looks like the variables LSTAT, RM, and PTRATIO are the most correlated to the sale price of the house.  These features might be used to create a basic predictor for the house sale price.

```python
features = list(data.columns)
features.remove('price')
```

```python
n_col = 3
n_row = 1 + np.ceil(len(features)/n_col)
plt.figure(figsize=(4*n_col,3*n_row))

for index, feature in enumerate(features):
    plt.subplot(n_row, n_col, index+1)
    plt.scatter(data[feature], data['price'], alpha=0.1, color='red')
    plt.xlabel(feature)
    plt.ylabel('price')
    
plt.tight_layout()
```

From the above we can see the approximate linear relationship between the LSTAT and the price, as well as the number of rooms and the sale price.  We might be encouraged by this to try creating a linear model that takes the LSTAT and RM variables as inputs and predicts the house price as an output.

## Modeling 
Having identified some basic features of the dataset, we can start trying to predict house prices based on the information we have. 

```python
baseline_features = ['RM', 'LSTAT', 'PTRATIO']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(data[baseline_features], data.price, test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import ElasticNet
from sklearn.linear_model import LassoCV
from sklearn.ensemble import RandomForestRegressor
```

```python
models = {}
models['linear'] = LinearRegression(n_jobs=-1)
models['elastic'] = ElasticNet()
models['lasso'] = LassoCV(n_jobs=-1)
models['random_forest'] = RandomForestRegressor(n_jobs=-1, n_estimators=10, max_depth=5)
```

```python
for m in models.iterkeys():
    print('training: %s' % m)
    models[m].fit(x_train, y_train)
```

```python
y_pred = {}

for name, model in models.iteritems():
    y_pred[name] = model.predict(x_test)
```

## Simple Metrics
Let's compare our models performance using some simple metrics.

```python
def r_square_metric(y_true, y_pred):
    return 1-np.sum((y_true-y_pred)**2)/np.sum(y_true**2)

def average_percent_error(y_true, y_pred):
    return np.average(np.abs((y_true-y_pred)/y_true))
```

```python
for name, model in models.iteritems():
    print('%s, r2_score=%.3f, mape=%.3f' % (name, r_square_metric(y_test, y_pred[name]), 
                                          average_percent_error(y_test, y_pred[name])))
```

We can visualize the results by scattering the prediciton for the test set against the true value.

```python
n_col = 2
n_row = np.ceil(len(models.keys())/n_col) + 1
plt.figure(figsize=(4*n_col, 3*n_row))

index = 0
for name, model in models.iteritems():
    index += 1
    
    plt.subplot(n_row, n_col, index)
    plt.scatter(y_test, y_pred[name], alpha=0.2, color='red')
    
    x = np.linspace(0, 50, 101)
    plt.plot(x, x, linestyle='--', alpha=0.5, color='black')

    plt.title(name.replace('_', ' '))
    plt.xlabel('true')
    plt.ylabel('predicted')
    
plt.tight_layout()
```

It's interesting to see that the linear models don't capture the correct behaviour of the housing prices for very expensive houses (they are predicting too low).  This might inspire one to look for a feature that is highly correlated with the price, at higher values of the price.

```python
datasub = data.query('price > 40')

n_col = 3
n_row = 1 + np.ceil(len(features)/n_col)
plt.figure(figsize=(4*n_col,3*n_row))

for index, feature in enumerate(features):
    plt.subplot(n_row, n_col, index+1)
    plt.scatter(datasub[feature], datasub['price'], alpha=0.1, color='red')
    plt.xlabel(feature)
    plt.ylabel('price')
    
plt.tight_layout()
```

```python
datasub.corr()['price']
```

The best variable to add seems to be DIS or CRIM.  Let's try adding both and re-training the linear models.

```python
better_features = ['RM', 'LSTAT', 'PTRATIO', 'DIS', 'CRIM']
```

```python
x_train, x_test, y_train, y_test = train_test_split(data[better_features], data.price, test_size=0.3)
```

```python
models = {}
models['linear'] = LinearRegression(n_jobs=-1)
models['elastic'] = ElasticNet()
models['lasso'] = LassoCV(n_jobs=-1)
models['random_forest'] = RandomForestRegressor(n_jobs=-1, n_estimators=10, max_depth=5)

for m in models.iterkeys():
    print('training: %s' % m)
    models[m].fit(x_train, y_train)

y_pred = {}
for name, model in models.iteritems():
    y_pred[name] = model.predict(x_test)
    
for name, model in models.iteritems():
    print('%s, r2_score=%.3f, mape=%.3f' % (name, r_square_metric(y_test, y_pred[name]), 
                                          average_percent_error(y_test, y_pred[name])))
```

We see that this added a slight boost to the performance of the linear models, as well as the RandomForest classifer.  This concludes this basic analysis.
